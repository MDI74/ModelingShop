```python
from multiprocessing import Queue
import random
from datetime import timedelta
import numpy as np
```

```python
def final_time(time, diff):
  return random.randint(time-diff, time+diff)
```

```python
class Que(): 
  def __init__(self, time, diff):
    self.time =time
    self.diff = diff
    self.final_time = 0

  def step(self):
    if self.final_time == 0:
      self.final_time = final_time(self.time, self.diff)
      if len(customer_list) < 20:
        customer_list.append(Customer(final_time(5*60, 3*60)))
    else:
      self.final_time-=1
      f = timedelta(seconds=self.final_time)
      print(f'Время до следующего покупателя {str(f)}')
```

```python
class Customer():
  def __init__(self, timer):
    self.time = timer
    self.wait_time = 0

  def step(self):
    if self.time:
      self.time -=1
      if len(customer_list) > 1:
        self.wait_time +=1
    else:
      wait_time_list.append(self.wait_time)
      self.wait_time = 0
      customer_list.remove(self)
      gen_time = final_time(5,5)
      if 1 < gen_time < 6:
        left_ready.append("ready")
      elif gen_time > 6 and gen_time < 10:
        right_ready.append("ready")
      else:
        leave.append("ready")
        print("Покупатель ушел")
```

```python
class Seller():
  def __init__(self, time, diff,flag):
    self.time = time + 60
    self.diff = diff
    self.final_time = 0
    self.place = "LEFT" if flag == 1 else "RIGHT"
    
  def step(self):
    if self.final_time == 0:
      if self.place == "LEFT":
        if left_ready:
          left_ready.pop()
          self.final_time = final_time(self.time, self.diff)
        else:
          print('Левая касса не работает')
          pass
      else:
        if right_ready:
          right_ready.pop()
          self.final_time = final_time(self.time, self.diff)
        else:
          print('Правая касса не работает')
          pass
    else:
      global check_right
      global check_left
      self.final_time -= 1
      if self.place == "LEFT":
        check_left +=1
      else:
         check_right +=1
      k = timedelta(seconds=self.final_time)
      print(f'Обслуживание покупателя {self.place} будет ещё {str(k)}')
```

```python
days = int(input('Количество дней:'))
hour = int(input('Количество часов:'))
come = int(input('Время прихода в секудах:'))
come_spread = int(input('Время разброса прихода в секудах:'))
check_left = 0
check_right = 0
wait_time_list = []
for day in range(1,days+1):
  come_time = Que(come, come_spread)
  seller_l = Seller(2*60, 60,1)
  seller_r = Seller(3*60, 2*60,0)
  left_ready = []
  right_ready = []
  leave = []
  customer_list = []
  for time in range(9*60*60,(9+hour)*60*60):
    t = timedelta(seconds=time)
    print('\nДень {0} Время {1} \n'.format(day, str(t)))
    print(f'Покупателей в магазине {len(customer_list)}')
    come_time.step()
    seller_l.step()
    seller_r.step()
    for x in customer_list:
      x.step()

print(f'\nСреднее время ожидания в очереди: {np.mean(np.array(wait_time_list))/60}')
print(f'Коэффициент загрузки левого кассира: {check_left/ (days*(hour)*60*60)}')
print(f'Коэффициент загрузки правого кассира: {check_right / (days*(hour)*60*60)}')
```
